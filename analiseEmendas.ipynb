```python
import pandas as pd
base = pd.read_csv('emendas_tratadas.csv', sep=';', encoding='utf-8')
```

```python
base['Data'].unique()
```

```python
base.columns
```

```python
# Tratamento da coluna 'Data'
base['Data'] = pd.to_datetime(base['Data'], format='%d/%m/%Y', errors='coerce')

# Tratamento da coluna 'Valor Pago'
base['Valor Pago'] = base['Valor Pago'].replace('.', '', regex=False)
base['Valor Pago'] = base['Valor Pago'].replace(',', '.', regex=False)
base['Valor Pago'] = base['Valor Pago'].replace('-', '', regex=False)
base['Valor Pago'] = base['Valor Pago'].replace('', '0')  # Substituir strings vazias por '0'
base['Valor Pago'] = base['Valor Pago'].astype(float)
```

```python
# filtrando Autor de Emenda para que tenha Bancada
base['Autor da Emenda'].unique()
```

```python
# selecionando os autores que tem a palavra BANCADA com contains
base_bancada = base[base['Autor da Emenda'].str.contains('BANCADA', na=False)]
base_bancada['Autor da Emenda'].unique()
```

```python
# Tratamento da coluna 'Data'
base_bancada['Data'] = pd.to_datetime(base_bancada['Data'], format='%d/%m/%Y', errors='coerce')

# Tratamento da coluna 'Valor Pago'
base_bancada['Valor Pago'] = base_bancada['Valor Pago'].replace('.', '', regex=False)
base_bancada['Valor Pago'] = base_bancada['Valor Pago'].replace(',', '.', regex=False)
base_bancada['Valor Pago'] = base_bancada['Valor Pago'].replace('-', '', regex=False)
base_bancada['Valor Pago'] = base_bancada['Valor Pago'].replace('', '0')  # Substituir strings vazias por '0'
base_bancada['Valor Pago'] = base_bancada['Valor Pago'].astype(float)
```

```python
# Agregação de Autor de Emenda Valor
base_bancada_agrupado = base_bancada.groupby('Autor da Emenda')['Valor Pago'].sum().reset_index()
base_bancada_agrupado = base_bancada_agrupado.sort_values(by='Valor Pago', ascending=False)
```

```python
# criando % de bancadas
base_bancada_agrupado['%'] = (base_bancada_agrupado['Valor Pago'] / base_bancada_agrupado['Valor Pago'].sum()) * 100
base_bancada_agrupado = base_bancada_agrupado.sort_values(by='%', ascending=False)
base_bancada_agrupado = base_bancada_agrupado.reset_index(drop=True)
base_bancada_agrupado['%'] = base_bancada_agrupado['%'].round(2)
base_bancada_agrupado.head(10)
```

```python
# Extraindo a base de bancada agregada
base_bancada_agrupado.to_csv('base_bancada_agrupada_2025.csv', sep=';', index=False, encoding='utf-8')
```

#### Total de Emendas por UF

```python
# emendas por uf
base['UF DO FAVORECIDO'].unique()
# criando agrupamento
base_uf = base.groupby('UF DO FAVORECIDO')['Valor Pago'].sum().reset_index()
base_uf = base_uf.sort_values(by='Valor Pago', ascending=False)
base_uf = base_uf.reset_index(drop=True)
base_uf['%'] = (base_uf['Valor Pago'] / base_uf['Valor Pago'].sum()) * 100
base_uf['%'] = base_uf['%'].round(2)
base_uf.head(10)
```

```python
# exportando a base de uf como emendas_uf_2025.csv
base_uf.to_csv('emendas_uf_2025.csv', sep=';', index=False, encoding='utf-8')
```

### Orgão superior

```python
sunburst_data = base.groupby([
    'Órgão', 
    'Ação Orçamentária', 
    'Função (Área de Atuação)', 
    'Subfunção (Especificação da Área de Atuação)'
]).agg({'Valor Pago': 'sum'}).reset_index()

# Tirando os 4 primeiros caracteres da coluna 'de subfunção
sunburst_data['Subfunção (Especificação da Área de Atuação)'] = sunburst_data['Subfunção (Especificação da Área de Atuação)'].str[6:]
sunburst_data['Subfunção (Especificação da Área de Atuação)'].unique()
# Em Função tambem
sunburst_data['Função (Área de Atuação)'] = sunburst_data['Função (Área de Atuação)'].str[5:]
sunburst_data['Função (Área de Atuação)'].unique()
# ação orçamentária
sunburst_data['Ação Orçamentária'] = sunburst_data['Ação Orçamentária'].str[6:]
sunburst_data['Ação Orçamentária'].unique()
```

```python
# exportando como emendasporOrgaoSUperior.csv
sunburst_data.to_csv('emendas_por_orgao_superior_2025.csv', sep=';', index=False, encoding='latin1')
```
